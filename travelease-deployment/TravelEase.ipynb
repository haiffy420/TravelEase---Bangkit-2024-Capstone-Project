# Data Modeling

TF-IDF Vectorizer

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
vectorizer = TfidfVectorizer()

vectorizer.fit(df['Category'])

vectorizer.get_feature_names_out()
```

```python
tfidf_matrix = vectorizer.fit_transform(df['Category'])
tfidf_matrix.shape
```

```python
# Convert TF-IDF matrix to a dense NumPy array
tfidf_matrix_dense = tfidf_matrix.toarray()
```

Split dataset

```python
X_train, X_test = train_test_split(tfidf_matrix_dense, test_size=0.2, random_state=42)
```

Model

```python
def build_model(input_dim):
    inputs = Input(shape=(input_dim,))
    x = Dense(128, activation='relu')(inputs)
    x = Dense(64, activation='relu')(x)
    outputs = Dense(input_dim, activation='softmax')(x)  # Using softmax for multi-class classification
    model = Model(inputs, outputs)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics='accuracy')
    return model
```

```python
# Custom callback to stop training when accuracy and val_accuracy reach 90%
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        if logs.get('accuracy') > 0.90 and logs.get('val_accuracy') > 0.90:
            print("\nReached 90% accuracy and validation accuracy, stopping training!")
            self.model.stop_training = True
```

```python
# Train the TensorFlow model
model = build_model(tfidf_matrix_dense.shape[1])
model.fit(X_train, X_train, epochs=100, batch_size=32, verbose=1, validation_data=(X_test, X_test), callbacks=[myCallback()])
```

Recommendation Testing

```python
def place_recommendations(categories, city, items=df[['Category', 'Place_Name', 'Rating', 'City']], k=5):
    # Filter dataframe based on specified city
    city_items = items[items['City'] == city]

    # If the number of items after filtering is less than k, adjust k
    if len(items) < k:
        k = len(items)

    # Transform input categories to TF-IDF vectors
    input_vector = vectorizer.transform(categories).toarray()

    # Predict TF-IDF representation of input categories using the trained model
    predicted_vector = model.predict(input_vector)

    # Compute cosine similarity between predicted vector and TF-IDF matrix for the city's items
    city_tfidf_matrix = vectorizer.transform(city_items['Category']).toarray()
    city_similarities = cosine_similarity(predicted_vector, city_tfidf_matrix)

    # Get indices of top similar places
    similar_indices = np.argsort(city_similarities, axis=1)[:, ::-1][:, :k]

    # Initialize an empty DataFrame to collect recommendations
    recommendations = pd.DataFrame(columns=['Place_Name', 'Category', 'Rating', 'City'])

    # Collect recommended places
    for i, indices in enumerate(similar_indices):
        category_places = city_items.iloc[indices]
        category_places['Category'] = categories[i]  # Assign the corresponding category to the recommended places
        recommendations = pd.concat([recommendations, category_places])

    # Sort recommendations based on ratings
    recommendations = recommendations.sort_values(by='Rating', ascending=False)

    return recommendations
```

```python
# Example usage
categories = ['Taman Hiburan', 'Budaya', 'Cagar Alam']
city = 'Yogyakarta'
print("Categories in original DataFrame:", df['Category'].unique())
```

```python
recommendations = place_recommendations(categories, city)
recommendations
```
